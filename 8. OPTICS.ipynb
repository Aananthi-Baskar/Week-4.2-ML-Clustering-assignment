```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X=dataset.iloc[:, [3,4]]
```

```python
X
```

```python
from sklearn.metrics.pairwise import euclidean_distances

distance = -euclidean_distances(X)

plt.imshow(distance)
plt.title("Euclidean Distance Matrix")
#plt.xlabel("Data Points")
#plt.ylabel('Euclidean distances')
#plt.colorbar()
#plt.show()
```

```python
from sklearn.cluster import OPTICS
```

```python
clusmodel = OPTICS(
    min_samples=11,                 # minimum points to form dense region
    max_eps=25,                   # maximum distance
    metric='euclidean',           # distance type
    cluster_method='xi',          # cluster extraction method
    xi=0.05,                      # sensitivity for cluster formation
    predecessor_correction=True,  # improves cluster structure
    algorithm='auto',             # neighbor search method
    leaf_size=30,                 # speed optimization
    n_jobs=-1                     # use all CPU cores
)
label=clusmodel.fit_predict(X)
```

```python
Supervised=pd.DataFrame(dataset)
```

```python
Supervised['cluster_group']=label
```

```python
Supervised
```

```python
import seaborn as sns
facet = sns.lmplot(
    data=Supervised,
    x=Supervised.columns[3],
    y=Supervised.columns[4],
    hue=Supervised.columns[5],
    fit_reg=False,
    legend=True,
    legend_out=True
)
```
