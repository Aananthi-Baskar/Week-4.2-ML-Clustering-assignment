```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv("Mall_Customers.csv")
```

```python
dataset
```

```python
X=dataset.iloc[:, [3,4]]
```

```python
X
```

```python
from sklearn.metrics.pairwise import euclidean_distances

distance = euclidean_distances(X)

plt.imshow(distance)
plt.title("Euclidean Distance Matrix")
plt.xlabel("Data Points")
plt.ylabel('Euclidean distances')
plt.colorbar()
plt.show()
```

```python
distance
```

```python
from sklearn.cluster import  estimate_bandwidth
bandwidth = estimate_bandwidth(X, quantile=0.2)
bandwidth
```

```python
from sklearn.cluster import MeanShift
```

```python
clusmodel= model = MeanShift(
    bandwidth=23,   # size of search window
    bin_seeding=False,     # faster initialization
    cluster_all=True,      # assign all points to clusters
    max_iter=300)

label=clusmodel.fit_predict(X)
```

```python
Supervised=pd.DataFrame(dataset)
```

```python
Supervised['cluster_group']=label
```

```python
Supervised
```

```python
import seaborn as sns
facet = sns.lmplot(
    data=Supervised,
    x=Supervised.columns[3],
    y=Supervised.columns[4],
    hue=Supervised.columns[5],
    fit_reg=False,
    legend=True,
    legend_out=True
)
```
